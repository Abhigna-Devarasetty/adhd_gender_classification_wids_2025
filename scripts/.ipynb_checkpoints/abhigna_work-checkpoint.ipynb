# WiDS 2025: ADHD Detection 

## About the Problem

### **1. Problem Statement**
The goal is to build a predictive model using functional brain imaging (fMRI) data, socio-demographic information, and questionnaire responses to:
1. Predict ADHD diagnosis (`ADHD_Outcome` - 1 for Yes, 0 for No).
2. Predict sex (`Sex_F` - 1 for Female, 0 for Male).


### **2. Dataset Overview**

#### **Dataset Folders**
1. **Train Folder (`train_tsv`):**
   - Contains data for 1,200+ subjects.
   - Includes:
     - **Targets**: ADHD diagnosis and sex.
     - **Functional MRI Connectome Matrices**.
     - **Socio-demographic Information**: 
       - Examples: handedness, parentâ€™s education level, emotions (Strength and Difficulties Questionnaire - SDQ), parenting information (Alabama Parenting Questionnaire - APQ).
     - Metadata: Includes both quantitative and categorical data.
   - Action Needed: Preprocess categorical data (e.g., create dummy variables) and combine with functional connectome dataset.

2. **Test Folder (`test_tsv`):**
   - Contains unseen data for 300+ subjects.
   - Includes:
     - Functional MRI Connectome Matrices.
     - Socio-demographic, emotions, and parenting information.
   - Goal: Use the test dataset to generate predictions for submission.

---

### **3. Key Variables**
1. **Target Variables:**
   - `ADHD_Outcome`: Type of Diagnosis (0 = Other/None, 1 = ADHD).
   - `Sex_F`: Sex of participant (0 = Male, 1 = Female).

2. **Metadata Variables:**
   - Includes a mix of numerical and categorical features related to demographics, parenting styles, and emotional health.

---

### **4. Evaluation Metric**
- The **F1 Score** is used as the evaluation metric:
  - It is the harmonic mean of precision and recall.
  - A weighted F1 Score is calculated for this competition, with **2x weight for Female ADHD cases** (`ADHD_Outcome=1` and `Sex_F=1`).
  - The final leaderboard score is the average of the weighted F1 Scores for the two target variables.
